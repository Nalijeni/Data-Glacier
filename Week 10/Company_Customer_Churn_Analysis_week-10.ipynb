##### 2.3.1.1 Age vs Total Spend with  Subscription Type

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x="Age", y="Total Spend", hue="Churn", data=df_new)

plt.title("Age vs Total Spend with Subscription Type")
plt.xlabel("Age")
plt.ylabel("Total Spend")
plt.legend()
plt.show()
```

#### 2.3.1.3 Usage Frequency, Contract Length, Last Interaction, Support Calls

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9, 8))
categories = ["Usage Frequency", "Contract Length", "Last Interaction", "Support Calls"]

for ax, category in zip(axes.flatten(), categories):
    sns.countplot(x=category, hue="Churn", data=df_new, palette="Set2", ax=ax)
    ax.set_title(f"Count of Subscription Type by {category.title()}")
    ax.set_xlabel(category.title())
    ax.set_ylabel("Count")
    ax.legend(title="Subscription Type", labels=["No", "Yes"])

plt.tight_layout()
plt.show()
```

The bar chart shows that while there are differences in subscription types acrossUsage Frequency, Contract Length, Last Interaction, Support Calls but not all have a great information.

```python
def categorical_visualization(col):
    plt.subplot(1,2,1)
    sns.countplot(x=col,data=df_new)
    plt.subplot(1,2,2)
    df_new[col].value_counts().plot(kind='pie',autopct='%0.2f%%')
    plt.tight_layout()
    plt.show()
    
    
def numerical_visualization(col):
    plt.subplot(1,2,1)
    sns.kdeplot(x=col,data=df_new)
    plt.grid()
    plt.subplot(1,2,2)
    sns.boxplot(x=col,data=df_new)
    plt.tight_layout()
    plt.show()
```

```python
for i in df.columns:
    if df_new[i].dtype=='object':
        categorical_visualization(i)
    else:
        numerical_visualization(i)
```
